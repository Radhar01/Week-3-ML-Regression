```python
#importing package pandas which have functions to work with database or other input data
import pandas as pd
```

```python
#Reading dataset from local computer path using read_csv function from pandas
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#printing the loaded dataset
dataset
```

```python
#Assigning Years of Experience column to variable named independent
independent=dataset[["YearsExperience"]]
```

```python
#Printing and validating the assigned variable and corresponding data
independent
```

```python
#Assigning Salary column to variable named dependent
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
#importing train_test_split from sklearn model selection package and splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Checking the no.of columns and rows of training data 
X_train.shape
```

```python
#Reading the actual training data for years of experience
X_train
```

```python
#importing linear regression function from sklearn package and linear_model
from sklearn.linear_model import LinearRegression
```

```python
#Assigning all the functions to a variable named regressor
regressor=LinearRegression()
```

```python
#Fitting the model to training data
regressor.fit(X_train,y_train)
```

```python
#Calculating the weightage using coef_ function
weight=regressor.coef_
```

```python
#Printing the weightage value
weight
```

```python
#Identifying the bias value using intercept_function
bias=regressor.intercept_
```

```python
#Printing the bias value
print(bias)
```

```python
#Predicting the salary values baesd on test data input
y_pred=regressor.predict(X_test)
```

```python
#Printing all the predicted values based on X_test input
y_pred
```

```python
#Printing and checking the y_test values once from the file
y_test
```

```python
#importing the r2_score function from sklearn metrics
from sklearn.metrics import r2_score
rscore=r2_score(y_test,y_pred)
```

```python
#Printing and check the rscore value based on y_pred and y_test actual values
rscore
```

```python
#Saving the model using pickle as the final model
import pickle
filename="finalModel.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Opening the saved model again using laod function
load_model=pickle.load(open("finalModel.sav",'rb'))
```

```python
#testing the saved model again using predict function
result=load_model.predict([[15]])
```
